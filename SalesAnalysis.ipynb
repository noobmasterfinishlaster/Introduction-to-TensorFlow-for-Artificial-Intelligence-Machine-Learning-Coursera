```python
df = pd.read_csv("all_data.csv")
df
```

**Extracting Month Column**

```python
nan_df = df[df.isna().any(axis= 1)]
nan_df.head()
```

**Drop rows of NAN**

```python
df = df.dropna(how = "any")
df.head()
```

```python
df =df[df['Order Date'].str[0:2] != 'Or']
df
```

**Addig month columnt**

```python
df['Month'] = df['Order Date'].str[0:2].astype('int32')
```

```python
df ['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df ['Price Each'] = pd.to_numeric(df['Price Each'])
```

**adding Sales COlum**

```python
df["Sales"] = df['Quantity Ordered'] * df ['Price Each']
df
```

**Finding Best Sales Months**

```python
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()
monthly_sales

```

```python
from matplotlib import pyplot as plt
monthly_sales['Sales'].plot(kind='line', figsize=(8, 4), title='Sales')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
monthly_summary = df.groupby('Month').agg({
    'Sales': 'sum',
    'Quantity Ordered': 'sum'
}).reset_index()
monthly_summary
```

```python
import matplotlib.pyplot as plt
months = range (1,13)
plt.bar(months,monthly_summary ['Sales'])
plt.xticks(months)
plt.ylabel ('Sales In Usd')
plt.xlabel ('Month Number')

plt.show()

```

```python
def get_city (address):
  return address.split(',')[1]
def get_states (address):
  return address.split(',')[2].split(' ')[1]


df ['City'] = df ['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_states(x)})")
df
```

```python
City_summary = df.groupby('City').agg({
    'Sales': 'sum',
    'Quantity Ordered': 'sum'
}).reset_index()
City_summary
```

```python
import matplotlib.pyplot as plt
citites = [city for city, df in df.groupby('City')]
plt.bar(citites,City_summary ['Sales'])
plt.xticks(citites, rotation = 'vertical',size = 8 )
plt.ylabel ('Sales In Usd')
plt.xlabel ('City Name ')

plt.show()
```

**What products are most often sold together?**

```python
df1 = df[df['Order ID'].duplicated(keep = False)]
df1['Grouped'] = df1.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df1 = df[df['Order ID'].duplicated(keep = False)]
df1['Grouped'] = df1.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df1[['Order ID', 'Grouped']].drop_duplicates()
df2
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

```python
product_group = df.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = [pair for pair, df in product_group]
plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
prices = df.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
